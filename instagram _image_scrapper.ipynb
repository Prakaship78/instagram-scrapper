```python
print("""\
       INSTAGRAM PHOTOS SCRAPPER "
   You can scrap all images of any of your friend on instagram 
   """)
a = input("Enter the friend username  ")
# open chrome
from bs4 import BeautifulSoup
import requests
from time import sleep
from selenium import webdriver
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import io

driver = webdriver.Chrome('C:\chromedriver')
insta_url = 'https://www.instagram.com/'


#open instagram and click on login button
driver.get("https://www.instagram.com/")
login = driver.find_element_by_xpath("//span[@id='react-root']//p[@class='izU2O']/a")
login.click()
sleep(3)
#fill username and password 
user = driver.find_element_by_name("username")
user.send_keys('prakaship78')
pswd = driver.find_element_by_name("password")
pswd.send_keys('12cl')

#after filling user and psw click on log in 

login_button = driver.find_element_by_xpath("//button[@class='_0mzm- sqdOP  L3NKy       ']")
login_button.click()
sleep(3)
# click on not now 
not_now = driver.find_element_by_xpath("//button[@class='aOOlW   HoLwm ']")
not_now.click()
sleep(1)

#opening a target user profile


target_username = driver.find_element_by_xpath("//input[@class='XTCLo x3qfX ']")
target_username.send_keys(a)
target_link = insta_url  + a +'/' 
driver.get(target_link) 


# No of posts
scroll = driver.find_element_by_xpath("//span[@class='g47SY ']")
scroll = str(scroll.text).replace(",","") #.replace to replace , sign with nthing
posts = int(scroll)
print(posts)

# finding no os scrolls

if posts > 12:
    scrolls = int(posts/12) 
    
    for i in range(scrolls):
        print(i)
        
# window.scrollTo(0), document.body.scrollHeight is in js and .executive_script is used to run the script
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(2)

# Extracting the img src link of all photos
soup = BeautifulSoup(driver.page_source,'lxml')
image = soup.find_all('img')
k = 1
for img in image:
    url= requests.get(img['src'],stream = True)
    img = Image.open(io.BytesIO(url.content))
    plt.imsave(arr=np.array(img),fname=str(k)+'.jpg')
    k +=1


```
